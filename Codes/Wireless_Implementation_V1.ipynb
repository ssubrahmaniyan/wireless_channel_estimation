```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

```python
# Generate 5 samples of Gaussian random numbers with mean 0 and standard deviation 1
gaussian_samples = np.random.normal(0, 1, 5)
print(gaussian_samples)
```

```python
#This set of numbers can be treated as a stochastic process with 5 samples
#The spectral characteristics of the process are given by the Power Spectral Density
#First we try to get the Fourier Transform of the sequence itsepf
fourier_transform = np.fft.fft(gaussian_samples)
print(fourier_transform)
#Real signal gives a conjugate symmetric fourier transform 
```

```python
magnitude = np.abs(fourier_transform)
phase = np.angle(fourier_transform)
```

```python
#Some quantities for discrete time Fourier Transform
N = len(gaussian_samples)
fs = 1 #Sampling rate
freq = np.fft.fftfreq(N, 1/fs)
```

```python
plt.figure(figsize = (12,6))

#Plot the magnitude
plt.subplot(121)
plt.vlines(freq, ymin=0, ymax=magnitude, colors='b', linestyles='solid')
plt.axhline(y=0, color='black', linestyle='--', linewidth=0.8)
plt.title('Magnitude of FFT')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')

# Plot the phase
plt.subplot(122)
plt.vlines(freq, ymin=0, ymax=phase, colors='r', linestyles='solid')
plt.axhline(y=0, color='black', linestyle='--', linewidth=0.8)
plt.title('Phase of FFT')
plt.xlabel('Frequency')
plt.ylabel('Phase (radians)')
```

Step 1  - 2 Correlated Gaussian Random Variables

```python
num1 = 40 #Number of samples of 1 standard Gaussian variable
X = gaussian_samples = np.random.normal(0, 1, num1) #Sample set 1 - variable 1 (mean, variance, num1)
Z = gaussian_samples = np.random.normal(0, 1, num1) #Dummy variable
```

```python
rho = 0.01 #Normalised correlation coefficient
```

```python
Y = rho*X + ((1-rho*rho)**0.5)*Z
```

```python
print(X)
print(Z)
print(Y)
```

```python
plt.figure(figsize = (12,6))
plt.scatter(Z, Y)
plt.title('2 Correlated Gaussian Variables')
```

Step 3 - Multiple Correlated Gaussian Variables or Correlated Gaussian Vector

```python
num2 = 2 #Number of variables to simulate, not the number of samples
```

```python
#Cholesky decomposition is done using eigenmatrices
mean_vector = [1, 1]
cov_matrix = [[1, 0.89],
              [0.89, 1]]
```

```python
#Standard Gaussian Random vector
#np.random.normal gives a random variable sample in a list, not as a float
Y = []
for i in range(num2):
    Y.append(float(np.random.normal(0, 1, 1)))
```

```python
print(Y)
```

```python
#Determining matrix B
eigenvalues, eigenmatrix = np.linalg.eig(cov_matrix)
B = eigenmatrix @ np.diag(np.sqrt(eigenvalues))
```

```python
#Matrix X with 2 correlated variables
X = mean_vector + B @ Y
print(X)
```

```python
X1_plot = []
X2_plot = []
num3 = 40 #Number of samples of the num2 number of variables for plotting
for i in range(num3):
    Y = []
    for i in range(num2):
        Y.append(float(np.random.normal(0, 1, 1)))
    eigenvalues, eigenmatrix = np.linalg.eig(cov_matrix)
    B = eigenmatrix @ np.diag(np.sqrt(eigenvalues))
    X = mean_vector + B @ Y
    X1_plot.append(X[0])
    X2_plot.append(X[1])

    
```

```python
plt.figure(figsize = (12,6))
plt.scatter(X1_plot, X2_plot)
plt.title('Multiple Correlated Non-Standard Gaussian Variables')
```

Step 4 - Correlated random processes with each process being a white Gaussian process

```python
#Consider 2 Processes W1 and W2 which are correspondingly correlated by a cross spectral density K12
#They have individual spectral densities K1 and K2
num1 = 40
K12 = 0.70 #Can vary this as rho since the other 2 spectral densities are taken as 1
K1 = 1 #Constant since the 2 processes are individually white
K2 = 1
delta_t = 0.01
rho2 = K12/((K1*K2)**0.5)
```

```python
W1 = []
W2 = []
for i in range(num1):
    X = float(np.random.normal(0, 1, 1))
    Z = float(np.random.normal(0, 1, 1))
    Y = rho2*X + ((1-rho2*rho2)**0.5)*Z
    W1.append(X*((2*math.pi*K1/delta_t)**0.5))
    W2.append(Y*((2*math.pi*K2/delta_t)**0.5))
```

```python
print(W1)
print(W2)
```

```python
plt.figure(figsize = (12,6))

plt.plot(W1, color = 'b', label = 'White Gaussian Process 1')
plt.plot(W2, color = 'r', label = 'White Gaussian Process 2')
plt.xlabel('Timesteps')
plt.ylabel('Amplitude - Gaussian')
plt.title('2 Correlated White Gaussian Processes')
```
