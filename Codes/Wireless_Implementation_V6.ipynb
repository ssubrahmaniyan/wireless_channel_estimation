```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import wiener
```

```python
# Parameters
N = 100  # Number of samples
x_const = 5  # Constant signal value
noise_level = 1.0  # Standard deviation of AWGN
```

```python
#Constant signal x, scaled by a constant
x = np.ones(N) * x_const
```

```python
# AWGN
noise = np.random.normal(0, noise_level, N)
x_noisy = x + noise
```

```python
# Wiener Filter
x_wiener = wiener(x_noisy)
```

```python
# Some plots
plt.figure(figsize=(12, 10))

# Plot original signal x
plt.subplot(3, 1, 1)
plt.plot(x, label='Original Signal x', color='blue')

# Plot noisy signal x_noisy
plt.plot(x_noisy, label='Noisy Signal', color='orange')

# Plot Wiener filtered signal x_wiener
plt.plot(x_wiener, label='Wiener Filtered Signal', linestyle='--', color='purple')

plt.title('Original Signal, Noisy Signal, and Wiener Filtered Signal')
plt.xlabel('Sample Index')
plt.ylabel('Signal Value')
plt.legend()
plt.grid(True)

# Plot final approximated signal x_wiener
plt.subplot(3, 1, 2)
plt.plot(x, label='Original Signal x', color='blue')
plt.plot(x_wiener, label='Wiener Filtered Signal', linestyle='--', color='green')

plt.title('Original Signal and Wiener Filtered Signal')
plt.xlabel('Sample Index')
plt.ylabel('Signal Value')
plt.legend()
plt.grid(True)

# Plot residuals from the Wiener filtering
plt.subplot(3, 1, 3)
plt.plot(x - x_wiener, label='Residuals', color='red')

plt.title('Residuals from Wiener Filtering')
plt.xlabel('Sample Index')
plt.ylabel('Residual Value')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

We see that the least squares approach is far better than other filters in removing AWGN
