```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
N = 1000  # Number of samples
x_const = 5  # Constant signal value
noise_level = 1.0  # Standard deviation of AWGN
```

```python
#Constant signal x, scaled by a constant
x = np.ones(N) * x_const
```

```python
#AWGN
noise = np.random.normal(0, noise_level, N)
x_noisy = x + noise
```

```python
# Least Squares
A = np.ones((N, 1))  # Using Ax = b, where b is the noisy signal and x is the original signal we need to approximate
coefficients, residuals, _, _ = np.linalg.lstsq(A, x_noisy, rcond=None) 
x_approx = A.dot(coefficients)
```

```python
# Plotting
plt.figure(figsize=(10, 6))

# Plot original signal x
plt.plot(x, label='Original Signal x', color='blue')

# Plot noisy signal x_noisy
plt.plot(x_noisy, label='Noisy Signal', color='orange')

# Plot final approximated signal x_approx
plt.plot(x_approx, label='Approximated Signal', linestyle='--', color='red')

plt.title('Original Signal, Noisy Signal, and Approximated Signal')
plt.xlabel('Sample Index')
plt.ylabel('Signal Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (10, 6))
plt.plot(x - x_approx, color='green')
plt.title('Residuals from Least Squares Approximation')
plt.xlabel('Sample Index')
plt.ylabel('Residual')
```
